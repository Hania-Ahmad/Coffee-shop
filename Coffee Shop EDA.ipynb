```python
df.columns
```

```python
# highest profit days
highest = df.groupby(['Date'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False).head(5)
sns.barplot(x = 'Date',y = 'Profit',data = highest)
plt.title("Highest Profit days");
```

```python
# highest sales day
highest = df.groupby(['Date'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False).head(5)
sns.barplot(x = 'Date',y = 'Sales',data = highest)
plt.title("Highest sales days");
```

```python
# profit by State
regionalsale = df.groupby(['State'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False)
sns.barplot(x = 'Profit',y = 'State',data = regionalsale)
plt.title("Profit by Region")
plt.figure(figsize = (15,6));
```

```python
# sales by state
regionalsale = df.groupby(['State'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False)
sns.barplot(x = 'Sales',y = 'State',data = regionalsale)
plt.title("Sales by Region")
plt.figure(figsize = (15,6));
```

```python
# regional profit based on type
regionalsaletype = df.groupby(['State'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False)
sns.barplot(x = 'Profit',y = 'State',data = regionalsaletype,hue =df.Product_line )
plt.title("Profit by Region leaves vs beans")
plt.figure(figsize = (15,6));
```

```python
# regional sales based on type
regionalsaletype = df.groupby(['State'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False)
sns.barplot(x = 'Sales',y = 'State',data = regionalsaletype,hue =df.Product_line )
plt.title("Sales by Region leaves vs beans")
plt.figure(figsize = (15,6));
```

```python
# Profit by product
plt.figure(figsize = (22,6))
productsale = df.groupby(['Product'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False)
sns.barplot(x = 'Product',y = 'Profit',data = productsale)
plt.title("Product profits");
```

```python
# product sales comparison
plt.figure(figsize = (22,6))
productsale = df.groupby(['Product'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False)
sns.barplot(x = 'Product',y = 'Sales',data = productsale)
plt.title("Product Sales");
```

```python
# product profit by state
plt.figure(figsize = (19,6))
productsale = df.groupby(['Product'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False)
sns.barplot(x = 'Product',y = 'Profit',data = productsale,hue = df.State)
plt.title("Product profits by states");
```

```python
# product sales by state analysis
plt.figure(figsize = (19,6))
productsale = df.groupby(['Product'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False)
sns.barplot(x = 'Product',y = 'Sales',data = productsale,hue = df.State)
plt.title("Product sales by states");
```

```python
# product availability by state
x = df.Product
y = df.State
plt.figure(figsize = (18,6))
sns.scatterplot(data = df , x = x, y= y)
plt.title('Products by state');
```

```python
# product and their type
x = df.Product
y = df.Type
plt.figure(figsize = (20,6))
sns.scatterplot(data = df , x = x , y = y )
```

```python
# product type profit analysis
plt.figure(figsize = (6,4))
producttypesale = df.groupby(['Product_type'],as_index = False)['Profit'].sum().sort_values(by = 'Profit',ascending = False)
sns.barplot(x = 'Product_type',y = 'Profit',data = producttypesale)
plt.title("Product types profits");
```

```python
# product type sales analysis
plt.figure(figsize = (6,4))
producttypesale = df.groupby(['Product_type'],as_index = False)['Sales'].sum().sort_values(by = 'Sales',ascending = False)
sns.barplot(x = 'Product_type',y = 'Sales',data = producttypesale)
plt.title("Product types sales");
```
